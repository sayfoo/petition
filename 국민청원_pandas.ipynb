```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
from urllib.parse import quote_plus
from selenium import webdriver
import pandas as pd
import time

driver = webdriver.Chrome()

end_page = 31 # 2019년 12월 3일 오후 6시 기준
index = 0 
columns = ["번호", "분류", "제목", "청원 종료일", "참여인원"]
df_petition = pd.DataFrame(columns=columns)

for page in range(1, end_page+1):
    # 국민청원 url
    url = f"https://www1.president.go.kr/petitions/?c=0&only=1&page={page}&order=1"
    driver.get(url)
    #driver.implicitly_wait(5)
    time.sleep(3)

    html = driver.page_source
    soup = BeautifulSoup(html, "html.parser")
    petition_table = soup.find("ul", class_="petition_list")
    
    for one_column in petition_table.find_all("div", class_="bl_wrap"):
        number = one_column.find("div", class_="bl_no").text.replace("번호 ", "")
        category = one_column.find("div", class_="bl_category ccategory cs wv_category").text.replace("분류 ", "")
        subject = one_column.find("div", class_="bl_subject").text.replace("제목 ", "")
        end_date = one_column.find("div", class_="bl_date light").text.replace("청원 종료일 ", "")
        agree = one_column.find("div", class_="bl_agree cs").text.replace("참여인원 ", "")
        
        df_petition.loc[index, "번호"] = number
        df_petition.loc[index, "분류"] = category
        df_petition.loc[index, "제목"] = subject
        df_petition.loc[index, "청원 종료일"] = end_date
        df_petition.loc[index, "참여인원"] = agree
        index += 1

        print("-"*80)
        print(f"{number}\n{category}\n{subject}\n{end_date}\n{agree}")
        
driver.close()
```
